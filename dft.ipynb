```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(sinal):
    N = len(sinal)
    X = np.zeros(N, dtype=complex)
    
    for k in range(N):
        X[k] = sum(sinal[n] * np.exp(-2j * np.pi * k * n / N) for n in range(N))
    
    return X
```

```python
histograma = [0, 0, 0, 1, 3, 4, 3, 2, 1, 0, 0, 0]
```

```python
totalAmostras = sum(histograma)
probabilidades = [contador / totalAmostras for contador in histograma]
```

```python
dftResultado = dft(probabilidades)
```

```python
plt.figure(figsize=(12, 4))

# histograma de frequências
plt.subplot(121)
plt.bar(range(len(histograma)), histograma, width=1, align='center')
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.title("Histograma de Frequências")

# dft das probabilidades
plt.subplot(122)
plt.plot(np.abs(dftResultado))
plt.xlabel("Frequência")
plt.ylabel("Magnitude")
plt.title("DFT do Histograma de Probabilidades")

plt.tight_layout()
plt.show()
```
