```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagemOriginal = cv2.imread('malha.jpg') 

# passando  para escala de cinza
imagemCinza = cv2.cvtColor(imagemOriginal, cv2.COLOR_BGR2GRAY)
```

```python
# imagem cinza

plt.imshow(imagemCinza, cmap='gray')
plt.title('Imagem Cinza'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# kernel gaussiano

tamanhoK= 5  # 5x5

sigma = 1.5  # desvio padro
kernel = np.fromfunction(
    lambda x, y: (1/ (2 * np.pi * sigma**2)) * np.exp(-((x - (tamanhoK - 1) / 2)**2 + (y - (tamanhoK - 1) / 2)**2) / (2 * sigma**2)),
    (tamanhoK, tamanhoK)
)
kernel /= np.sum(kernel)  # normaliza
```

```python
# aplicando a filtragem
imagemFiltrada = cv2.filter2D(imagemCinza, -1, kernel)
```

```python
# imagem filtrada

plt.imshow(imagemFiltrada, cmap='gray')
plt.title('Imagem Filtrada'), plt.xticks([]), plt.yticks([])
plt.show()
```
